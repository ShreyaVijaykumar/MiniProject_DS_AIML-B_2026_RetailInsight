## 1. Environment Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
plt.style.use('dark_background')
sns.set_palette('husl')

print('Libraries loaded successfully.')
```

## 2. Load Raw Dataset

The Online Retail II dataset contains two Excel sheets — one per year. We load and concatenate both.

**Download first:** https://www.kaggle.com/datasets/mashlyn/online-retail-ii-uci  
Place the file at `../dataset/raw_data/online_retail_II.xlsx`

```python
RAW_PATH = '../dataset/raw_data/online_retail_II.csv'

df1 = pd.read_excel(RAW_PATH, sheet_name='Year 2009-2010', engine='openpyxl')
df2 = pd.read_excel(RAW_PATH, sheet_name='Year 2010-2011', engine='openpyxl')
df  = pd.concat([df1, df2], ignore_index=True)

print(f'Total rows loaded: {len(df):,}')
print(f'Sheet 1 (2009-2010): {len(df1):,} rows')
print(f'Sheet 2 (2010-2011): {len(df2):,} rows')
```

## 3. Dataset Structure

```python
print('Shape:', df.shape)
print('\nColumn Names:')
for col in df.columns:
    print(f'  - {col}: {df[col].dtype}')
```

```python
df.head(10)
```

## 4. Data Types & Missing Values

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
print('Missing Values:\n')
print(missing_df[missing_df['Missing Count'] > 0].to_string())
```

```python
# Visualise missing values as a heatmap
fig, ax = plt.subplots(figsize=(10, 4))
sns.heatmap(df.isnull(), yticklabels=False, cbar=True, cmap='YlOrRd', ax=ax)
ax.set_title('Missing Value Heatmap', fontsize=13, pad=10)
plt.tight_layout()
plt.savefig('../outputs/graphs/missing_values_heatmap.png', dpi=130, bbox_inches='tight')
plt.show()
```

## 6. Key Observations

Findings from initial inspection:

```python
# Cancellations (InvoiceNo starts with 'C')
cancellations = df[df['Invoice'].astype(str).str.startswith('C')]
print(f'Cancellation rows: {len(cancellations):,} ({len(cancellations)/len(df)*100:.2f}%)')

# Negative quantities
neg_qty = df[df['Quantity'] < 0]
print(f'Negative Quantity rows: {len(neg_qty):,}')

# Zero or negative prices
bad_price = df[df['Price'] <= 0]
print(f'Zero/Negative Price rows: {len(bad_price):,}')

# Unique counts
print(f'\nUnique Customers  : {df["Customer ID"].nunique():,}')
print(f'Unique Products   : {df["StockCode"].nunique():,}')
print(f'Unique Countries  : {df["Country"].nunique():,}')
print(f'Date Range        : {df["InvoiceDate"].min()} → {df["InvoiceDate"].max()}')
```

```python
# Top 10 countries by transaction count
top_countries = df['Country'].value_counts().head(10)

fig, ax = plt.subplots(figsize=(10, 5))
top_countries.plot(kind='barh', ax=ax, color='#4fc3f7', edgecolor='none')
ax.set_title('Top 10 Countries by Transaction Count', fontsize=13)
ax.set_xlabel('Transactions')
ax.invert_yaxis()
plt.tight_layout()
plt.savefig('../outputs/graphs/top_countries.png', dpi=130, bbox_inches='tight')
plt.show()
```
